```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given parameters
mean_lifetime = 1200  # Mean lifetime of the bulbs (hours)
std_dev = 250  # Standard deviation of the lifetime (hours)
sample_size = 60  # Number of bulbs in the sample
desired_lifetime = 1250  # Desired lifetime (hours)

# Calculate the mean and standard deviation of the sample distribution
mean_sample = mean_lifetime
std_dev_sample = std_dev / (sample_size ** 0.5)

# Calculate the Z-score for the desired lifetime
z_score = (desired_lifetime - mean_sample) / std_dev_sample

# Find the probability using the cumulative distribution function (CDF)
# We're finding the probability that the lifetime exceeds 1250 hours, so we use 1 - CDF
probability = 1 - stats.norm.cdf(z_score)

# Plotting the normal distribution and highlighting the area representing the probability
x = np.linspace(mean_lifetime - 3 * std_dev_sample, mean_lifetime + 3 * std_dev_sample, 1000)
y = stats.norm.pdf(x, mean_lifetime, std_dev_sample)

plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=(x > desired_lifetime), color='skyblue', alpha=0.5, label='Probability')

plt.title('Normal Distribution of Bulb Lifetime')
plt.xlabel('Lifetime (hours)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print("Probability that the average lifetime of 60 bulbs exceeds 1250 hrs:", probability)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given parameters
lambda_val = 2  # Poisson parameter lambda
n = 75  # Number of Poisson variates
lower_bound = 120
upper_bound = 160

# Calculate mean and standard deviation of Sn
mean_Sn = n * lambda_val
std_dev_Sn = np.sqrt(n * lambda_val)

# Calculate Z-scores for the lower and upper bounds
z_lower = (lower_bound - mean_Sn) / std_dev_Sn
z_upper = (upper_bound - mean_Sn) / std_dev_Sn

# Use the cumulative distribution function (CDF) to find the probability
# P(120 < Sn < 160) = P(Sn < 160) - P(Sn < 120)
probability = stats.norm.cdf(z_upper) - stats.norm.cdf(z_lower)

print("Estimated probability P(120 < Sn < 160):", probability)

# Plotting the normal distribution and shading the area representing the probability
x = np.linspace(mean_Sn - 3 * std_dev_Sn, mean_Sn + 3 * std_dev_Sn, 1000)
y = stats.norm.pdf(x, mean_Sn, std_dev_Sn)

plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=(x > lower_bound) & (x < upper_bound), color='skyblue', alpha=0.5, label='Probability')

plt.title('Normal Distribution of Sn')
plt.xlabel('Sn')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
